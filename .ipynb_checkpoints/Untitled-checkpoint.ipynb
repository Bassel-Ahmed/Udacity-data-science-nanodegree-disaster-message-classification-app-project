```python
%matplotlib inline
import pandas as pd
import numpy as np
#import matplotlib.pyplt as plt
```

```python
from sqlalchemy import create_engine
```

```python
categories=pd.read_csv('data/disaster_categories.csv')
messages=pd.read_csv('data/disaster_messages.csv')
```

```python
categories.sample(5)
```

```python
messages.sample(5)
```

```python
# to see if there is inconsistencies in coumn nmaes to be able merge the 2 dataframes
print(categories.columns)
print(messages.columns)
```

we alo need to check if there are any consistencies in values in the common column (id) to check that we can use info()method as it show the type and number of missing values

```python
categories.info()
messages.info()
```

```python
categories.drop_duplicates(inplace=True)
messages.drop_duplicates(inplace=True)
print(categories.shape)
print(messages.shape)
```

Now we are ready for mergign the 2 dataframes .
we see from samples above the id column is the primary key that identify each row of the data 

```python
data=pd.merge(left=messages,right=categories,on ='id',how='inner')
data.sample(5)
```

```python
print(messages.shape)
print(categories.shape)
print(data.shape)
```

we see from the information about messages dataframe that column original has many nan values specifcally  more than the half  
So it's beeter to drop it 

```python
data.drop(['original'],axis=1,inplace =True)
```

```python
data.info()
```

Now we have no missing values in data now we nedd to separate values in catgeory column into 26 columns t 
Also adjust types column 
Genre column msut be of category type


```python
data['genre'].astype('category')
data['genre'].unique()
```

```python
data['categories'][15280]
```

```python
# to get the names of the columns 
names=[]
l=data['categories'][15].strip().split(';')
for name in l:
    names.append(name.split('-')[0])
classes=data['categories'].str.split(';',expand=True)
classes.columns=names
classes
```

```python
for name in names:
    classes[name]=classes[name].apply(lambda x: x.split('-')[1])
classes
```

we need to modify the type of each column

```python
for column in classes.columns:
    classes[column]=pd.to_numeric(classes[column])
```

we need to be sure that the number of unique values in each columns is just 2(0,1)

```python
for name in classes.columns:
    if len(classes[name].unique())!=2:
        print(name+str(classes[name].unique()))
    
```

We see that the child_ alone column has no value other than 0 so for predicting it we always predict 0 and this doesn't give any information and doesn't need any model for prediction
Alo column related has vlaue of 2 we need to see the messages of this value and see what they express
but before that we nedd to concatenate all catgeories to data dataframe 

```python
classes['related'].value_counts()/classes.shape[0]
```

```python

df=pd.concat([data,classes],axis=1)
df.sample(4)
```

```python
df[df['related']==2]['message']
```

We see that most of the messages of related column has value of are in different languages that not useful at building machine learning model so we will drop them
Also drop categories column as we resolved the categories in it and child_alone column

```python
df=df[df['related']!=2]
df.drop(['categories','child_alone'],axis=1,inplace=True)
```

```python
engine = create_engine('sqlite:///Messages.db')
df.to_sql('Messages', engine, index=False, if_exists='replace')
```
